 1.Initialize centroid

 2.Compute the nearest data points  distance from the centroid
  Euclidean distance-small dimension data
  Manhottan distance-High dimension data,parse data (data points are far from one another)

 3.Draw  the perpendiculars and find the average of nearest data point distance to centroids

 4. Move the centroid to the average points calculated above

 5.Again repeat s-2 to s-4 until the current average points are similar to the last centroid points

 6.We will get the centroid points of the data points and draw the clusters.
